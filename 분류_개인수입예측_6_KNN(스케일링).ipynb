### 빼놓은 income을 y에 넣기

```python
y = targets
```

## 특성 중요도
- onehot_train : 결측치 처리(컬럼삭제, 행 1개삭제, 널값채우기) 후 원핫인코딩한 train
- onehot_test
- 랜덤포레스트회귀 모델
- 나무100개, random_state=7

```python
from sklearn.ensemble import RandomForestClassifier

ranfo_classifier = RandomForestClassifier(n_estimators=200, random_state=7)
ranfo_classifier.fit(onehot_train, y)

fi = ranfo_classifier.feature_importances_
importance_df = pd.DataFrame(fi, index=onehot_train.columns)
importance_df.sort_values(by=0, ascending=False)
```

## 특성 선택

```python
from sklearn.feature_selection import SelectFromModel                  # 특성선택해주는 라이브러리
selectModel = SelectFromModel(ranfo_classifier, prefit=True, threshold=0.00001)
train_reduced = selectModel.transform(onehot_train)
test_reduced = selectModel.transform(onehot_test)
print(train_reduced.shape)
print(test_reduced.shape)
```

### X_train, X_test, y_train, y_test

```python
from sklearn.model_selection import train_test_split                        
X_train, X_test, y_train, y_test = train_test_split(train_reduced, y, random_state=7) 
```

### 스케일링

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
scaler.fit(X_train)    # 스케일은 test로 하면 안된다. train으로만 스케일링 모델을 만든다.
X_train_transform = scaler.transform(X_train)
X_test_transform = scaler.transform(X_test)
```
