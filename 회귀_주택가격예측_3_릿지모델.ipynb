## 원핫인코딩 후 Trina 과 Test 나누기

```python
onehot_train = onehot_combined.iloc[:1750, :]
onehot_test = onehot_combined.iloc[1750:, :]
```

```python
print(onehot_train.shape)
print(onehot_test.shape)
```

## 특성 중요도
- onehot_train : 결측치 처리(컬럼삭제, 행 1개삭제, 널값채우기) 후 원핫인코딩한 train
- onehot_test
- 랜덤포레스트회귀 모델
- 나무100개, random_state=7

```python
from sklearn.ensemble import RandomForestRegressor

ranfo_reg = RandomForestRegressor(n_estimators=200, random_state=7)
ranfo_reg.fit(onehot_train, y)

fi = ranfo_reg.feature_importances_
importance_df = pd.DataFrame(fi, index=onehot_train.columns)
importance_df.sort_values(by=0, ascending=False)
```

- 소수점예시 : 1.990832e-02 : 0.01990832

### 특성 선택

```python
from sklearn.feature_selection import SelectFromModel                  # 특성선택해주는 라이브러리
selectModel = SelectFromModel(ranfo_reg, prefit=True, threshold=0.00007)
train_reduced = selectModel.transform(onehot_train)
test_reduced = selectModel.transform(onehot_test)
print(train_reduced.shape)
print(test_reduced.shape)
```

## Ridge 모델

```python
from sklearn.model_selection import train_test_split                        
X_train, X_test, y_train, y_test = train_test_split(train_reduced, y, random_state=7) 

from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score

for a in range(1, 50):
    ridge = Ridge(alpha=a)
    result_ridge = cross_val_score(ridge, X_train, y_train, cv=5)
    print(a, result_ridge.mean())
```

alpha=32

## 학습하기

```python
final_ridge = Ridge(alpha=11)
final_ridge.fit(X_train, y_train)
print(final_ridge.score(X_train, y_train))
```

## 평가하기

```python
final_ridge.fit(X_test, y_test)
print(final_ridge.score(X_test, y_test))
```

## 예측하기

```python
pre = final_ridge.predict(test_reduced)
```
